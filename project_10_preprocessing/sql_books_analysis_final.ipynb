##  - Для каждой книги посчитайте количество обзоров и среднюю оценку;

```
# вычислим ко-во обзоров и среднюю оценку, отсортируем сначало по кол-ву обзоров
number_of_reviews = '''
SELECT books.title AS book_title,
    AVG (ratings.rating) AS avg_rating,
    COUNT (DISTINCT reviews.review_id) AS cnt_review
FROM books 
    INNER JOIN ratings ON ratings.book_id = books.book_id
    LEFT JOIN reviews ON reviews.book_id = books.book_id
GROUP BY books.book_id
ORDER BY cnt_review DESC, avg_rating DESC
'''

number_of_reviews = pd.io.sql.read_sql(number_of_reviews, con = engine)
number_of_reviews
```

```
# вычислим ко-во обзоров и среднюю оценку, отсортируем сначало по средней оценке
number_of_reviews2 = '''
SELECT books.title AS book_title,
    AVG (ratings.rating) AS avg_rating,
    COUNT (DISTINCT reviews.review_id) AS cnt_review
FROM books 
    INNER JOIN ratings ON ratings.book_id = books.book_id
    LEFT JOIN reviews ON reviews.book_id = books.book_id
GROUP BY books.book_id
ORDER BY avg_rating DESC, cnt_review DESC
'''

number_of_reviews2 = pd.io.sql.read_sql(number_of_reviews2, con = engine)
number_of_reviews2
```

##   - Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```
# вычислим издательство 
books_thicker_50 = '''
SELECT publisher AS publisher_name,
    COUNT (books.book_id) AS cnt_books
FROM books 
    INNER JOIN publishers ON books.publisher_id = publishers.publisher_id
WHERE books.num_pages > 50
GROUP  BY publisher
ORDER  BY cnt_books DESC
LIMIT 10
'''

books_thicker_50 = pd.io.sql.read_sql(books_thicker_50, con = engine)
books_thicker_50
```

### Вывод:
- издательство, которое выпустило наибольшее число книг толще 50 страниц - Penguin Books

##    - Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```
# вычислим автора с самой высокой средней оценкой
author_highest = '''
SELECT authors.author AS author_name,
AVG(r.average_rating) as avg_rating
FROM books 
INNER JOIN (SELECT book_id, COUNT(rating) AS cnt,
             AVG(rating) AS average_rating
      FROM ratings
      GROUP BY book_id) AS r
ON r.book_id = books.book_id
INNER JOIN authors ON authors.author_id = books.author_id
WHERE r.cnt >= 50
GROUP BY author_name
ORDER BY avg_rating DESC
LIMIT 10
'''
author_highest = pd.io.sql.read_sql(author_highest, con = engine)
author_highest
```

### Вывод:
- автор с самой высокой средней оценкой книг 4,28 - J.K. Rowling/Mary GrandPré - написавшая Гарри Поттер

##   - Посчитайте среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```
# вычеслим среднее количество обзоров от пользователей
number_user_reviews = '''
SELECT
AVG(sub.cnt_text)
FROM
(SELECT 
   reviews.username,
   COUNT(DISTINCT reviews.text) AS cnt_text,
   COUNT(DISTINCT ratings.book_id) AS cnt_book
FROM reviews
     INNER JOIN ratings ON ratings.username = reviews.username
GROUP BY
   reviews.username
HAVING
   COUNT(DISTINCT ratings.book_id)>48) AS sub
'''
number_user_reviews = pd.io.sql.read_sql(number_user_reviews, con = engine)
number_user_reviews
```

### Вывод:
- среднее количество обзоров от пользователей, которые поставили больше 48 оценок равно 24 обзорам
